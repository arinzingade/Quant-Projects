```python
import yfinance as yf
import os
from dotenv import load_dotenv
load_dotenv()
from renko import build_renko_with_stocktrends
from datetime import datetime
```

```python
df = yf.download('BTC-USD', start = "2025-04-4", end = "2025-04-12", interval="1m")
```

```python
df.columns = df.columns.droplevel(1)
```

```python
df = df.reset_index()

df.rename(columns={
    'Datetime': 'date',
    'Open': 'open',
    'High': 'high',
    'Low': 'low',
    'Close': 'close',
    'Adj Close': 'adjclose',
    'Volume': 'volume'
}, inplace=True)


```

```python
import pandas as pd
pd.set_option('display.max_rows', None)

# Display all columns
pd.set_option('display.max_columns', None)

# Avoid column width truncation
pd.set_option('display.max_colwidth', None)
```

```python
bullish_instances = 0
profit = 0

win = 0
optimal_brick_size = 0
qty = 0.001
fees = qty * 85000 * 0.0005

combos = {}

for brick in range(100, 750):
    renko_df = build_renko_with_stocktrends(df, brick)

    bullish_instances = 0
    profit = 0
    pnl = 0

    for i in range(2, len(renko_df) - 4):
        third_last = renko_df['uptrend'].iloc[i - 2]
        second_last = renko_df['uptrend'].iloc[i - 1]
        last = renko_df['uptrend'].iloc[i]

        if second_last == False and last == True:
            bullish_instances += 1

            forward = renko_df['uptrend'].iloc[i + 1]
            twice_forward = renko_df['uptrend'].iloc[i+2]

            if forward == True:
                profit += 1

    if bullish_instances > 0:
        win_here = profit / bullish_instances

        if win_here > win:
            win = win_here
            optimal_brick_size = brick
        
        pnl += (profit) * (brick) - (bullish_instances - profit) * brick 
        print(pnl * qty, bullish_instances*fees*2 ,win_here, brick)

        combos[brick] = win_here
```

```python
pd.DataFrame(list(combos.items()), columns=["brick_size", "win_rate"])
```
